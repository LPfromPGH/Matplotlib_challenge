### Analysis

- 1. Larger mice have larger tumors. While there may be a thought process of a smaller mouse having a higher metabloism and therefore more aggresive tumors, it would seem that having more mass means having more mass that can turn into tumor.
- 2. Capomulin and Ramicane were clearly the best medications for shrinking tumor volume. They both had an average tumor size in the 40.2 - 40.6 range. They both also had the hoghest number of timepoints, which would indicate mice on those drugs survived longer than the mice on the other options.
- 3. While the correlation coefficient is only .53, the linear regression clearly shows that there is a connection between mice size and tumor size. The next round of testing it might make sense to try and get mice that are closer in weight.
 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame

allmice = pd.merge(mouse_metadata, study_results, on= "Mouse ID")

# Display the data table for preview
allmice
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.

micebymed = clnallmice.groupby(["Drug Regimen"])

```

```python
avgtumr = micebymed["Tumor Volume (mm3)"].mean()
meditumr = micebymed["Tumor Volume (mm3)"].median()
varitumr = micebymed["Tumor Volume (mm3)"].var()
stdtumr = micebymed["Tumor Volume (mm3)"].std()
semtumr = micebymed["Tumor Volume (mm3)"].sem()

tumornumbr = pd.DataFrame({"Avg Tumor Size": avgtumr, "Median Tumor Size": meditumr, "Variance in Tumor Size": varitumr,"STD Deviation in Tumor Size": stdtumr, "SEM in Tumor Size": semtumr})
tumornumbr.head(10)
```

```python
clnallmice.info()
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
miceaggr = clnallmice.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].agg(['mean', 'median', 'var', 'std', 'sem'])
miceaggr
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
capomice = mrgallmice[["Mouse ID", "Drug Regimen", "Weight (g)", "Tumor Volume (mm3)"]]
capomice = capomice.sort_values(["Mouse ID", "Weight (g)"])
capomice = capomice[capomice["Drug Regimen"] == "Capomulin"]
capomice.plot(kind="scatter", x="Weight (g)", y="Tumor Volume (mm3)", grid=True, figsize=(5,5), title="Weight Vs Tumor Volume")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
micemass = capomice["Weight (g)"]
tumormass = capomice["Tumor Volume (mm3)"]
print(f"The Correlation Coefficient between Mouse Weight and Tumor Volume is {round(st.pearsonr(micemass,tumormass)[0],2)}")
tmslope, tmint, tmr, tmp, tmstderr = st.linregress(micemass, tumormass)
tmfit = tmslope * micemass + tmint
capomice.plot(kind="scatter", x="Weight (g)", y=("Tumor Volume (mm3)"), grid=True, figsize=(8,8), title="Weight Vs Tumor Volume")
plt.plot(micemass,tmfit,"--")
plt.show()
plt.show()
```
